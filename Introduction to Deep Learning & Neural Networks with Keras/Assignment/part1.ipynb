```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('concrete_data.csv')
df.describe()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# X_train, Y_train
# X_test, Y_test
```

```python
df_columns = df.columns
predictors = df[df_columns[df_columns!='Strength']]
target = df['Strength']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(predictors, target, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
n_cols = predictors.shape[1] # number of predictors
n_cols
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# build the model
model = regression_model()
```

```python
# fit the model
epochs = 50
model.fit(X_train, Y_train, epochs=epochs, verbose=1)
```

```python
loss_val = model.evaluate(X_test, Y_test)
Y_pred = model.predict(X_test)
loss_val
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_square_error = mean_squared_error(Y_test, Y_pred)
mean = np.mean(mean_square_error)
standard_deviation = np.std(mean_square_error)
print(mean, standard_deviation)
```

```python
# Creating the list

total_mean_squared_errors = 50
epochs = 50
mean_squared_errors = []
for i in range(0, total_mean_squared_errors):
    X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.3, random_state=i)
    model.fit(X_train, y_train, epochs=epochs, verbose=0)
    MSE = model.evaluate(X_test, y_test, verbose=0)
    print("MSE "+str(i+1)+": "+str(MSE))
    y_pred = model.predict(X_test)
    mean_square_error = mean_squared_error(y_test, y_pred)
    mean_squared_errors.append(mean_square_error)

mean_squared_errors = np.array(mean_squared_errors)
mean = np.mean(mean_squared_errors)
standard_deviation = np.std(mean_squared_errors)

print('\n')

print("Mean of 50 mean squared errors without normalized data: "+str(mean))
print("Standard Deviation of 50 mean squared errors without normalized data: "+str(standard_deviation))
```
